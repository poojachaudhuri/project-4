```python
# imports

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.nycbynatives.com/nyc_info/new_york_city_zip_codes.php'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# find all tr tags

table = soup.find_all('tr')
table
```

```python
# get the first tr tag

table[0]
```

```python
# get the first tr tag wothout the html tags

table[0].text
```

```python
# get the td tags

table[0].find_all('td')
```

```python
# get all the tr tags without the html tags and put them in a dataframe

df = pd.DataFrame([[table[i].find_all('td')[j].text for j in range(len(table[i].find_all('td')))] for i in range(len(table))])
df
```

```python
df.columns
```

```python
# remove column 2

df = df.drop(columns=2)
df
```

```python
# rename column 0 to zip

df.rename(columns={0:'zip', 1:'borough', 3:'zip1', 4:'borough2'}, inplace=True)
df
```

```python
# remove \t from the zip column

df['zip'] = df['zip'].str.replace('\t', '')
df
```

```python
# remove \t from all columns

df['borough'] = df['borough'].str.replace('\t', '')
df['borough2'] = df['borough2'].str.replace('\t', '')
df['zip1'] = df['zip1'].str.replace('\t', '')
df
```

```python
# reset the index

df.reset_index(drop=True, inplace=True)
df
```
